```
transaction_null = spark.read.format("csv").option("header","true").option("InferSchema","true")\
    .load("/FileStore/tables/cleanedup_null_transactions_data-1.csv")
transaction_null.printSchema()   
```

```
print("No. of rows :", transaction_null.count())
print("No. of columns :" ,len(transaction_null.columns))
```

```
transaction_null = transaction_null.dropDuplicates()   
```

```
print("No. of rows :", transaction_null.count())
print("No. of columns :" ,len(transaction_null.columns))
```

```
display(transaction_null)
```

```
transaction_null.createOrReplaceTempView("transaction_null")
```

```
%sql
select round(sum(ifnull(transactionAmount,0)),2) as Total_transaction_amount from transaction_null;
```

```
%sql
select round(avg(ifnull(transactionAmount,0)),2) as Total_transaction_amount from transaction_null;
```

```
from pyspark.sql.types import FloatType
transaction_null = transaction_null.withColumn("transactionAmount",
transaction_null["transactionAmount"].cast(FloatType()))  
```

```
%sql
SELECT merchantState,count(*) as Count FROM transaction_null GROUP BY merchantState ORDER  BY COUNT(*) DESC;
```

```
%sql
select distinct(merchantState) from transaction_null
```
